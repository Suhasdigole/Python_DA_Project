```python
# Importing all the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# importing the dataset
# Importing the required csv file
epl=pd.read_csv("EPL_20_21.csv")
```

```python
#Checking the null values
epl.isnull().sum()
```

```python
#Creating 2 more columns for further analysis
epl['MinsPerMatch']=(epl['Mins']/epl['Matches']).astype(int)
epl['GoalsPerMatch']=(epl['Goals']/epl['Matches']).astype(float)
```

```python
epl.head()
```

```python
#Calculating the total goals scored in the EPL
Total_goals= epl['Goals'].sum()
print(Total_goals)
```

```python
# Calculating the goals scored by penalty
Total_Penalty_Goals= epl['Penalty_Goals'].sum()
print(Total_Penalty_Goals)
```

```python
#Calculating the attempted penalties
Total_Penalty_Attempted= epl['Penalty_Attempted'].sum()
print(Total_Penalty_Attempted)
```

```python
# Percentge of the penalty attempted vs penalty scored
Percentage_of_PenaltyScored= (Total_Penalty_Goals/Total_Penalty_Attempted)
print(Percentage_of_PenaltyScored)
```

```python
#Percentage of penalties  not scored
Percentage_of_PenaltyNotScored=1 - Percentage_of_PenaltyScored
print(Percentage_of_PenaltyNotScored)
```

```python
#Pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
data=[Percentage_of_PenaltyScored,Percentage_of_PenaltyNotScored]
labels= ['Penalty scored','Penalty missed']
color=sns.color_palette('hls')
plt.pie(data,labels=labels,colors=color,autopct="%.0f%%")
plt.show()

```

```python
# Players mostly from which countries
nationality= epl.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("husl"))
```

```python
#Maximum no of players in the squad (TOP-5)
epl['Club'].value_counts().nlargest(5).plot(kind='bar',figsize=(12,6),color=sns.color_palette("viridis"))
```

```python
#Least no of players in the squad (TOP-5)
epl['Club'].value_counts().nsmallest(5).plot(kind='bar',figsize=(12,6),color=sns.color_palette("viridis"))
```

```python
#Players based on age group
Under_20 =epl[epl['Age']<=20]
age_20_25 = epl[(epl['Age'] >20) & (epl['Age'] <= 25)]
age_25_30 = epl[(epl['Age'] >25) & (epl['Age'] <=30 )]
Above30 = epl[epl['Age']>30]
```

```python
data1=np.array([Under_20['Name'].count(),age_20_25['Name'].count(),age_25_30['Name'].count(),Above30['Name'].count()])
mylabels=['<=20','>20 & <=25','>25 & <=30','>30']
plt.title('Top players with age',fontsize=20)
plt.pie(data1,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# Players under 20
players_under_20=epl[epl['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette('cubehelix'))
```

```python
#average age of players in each club
plt.figure(figsize=(15,6))
sns.boxplot(x="Club",y="Age",data=epl)
plt.xticks(rotation=90)
```

```python
# Above you can see the average age of players for each club and can determine the overall age of the clubs 
#i.e (how old or young the team is)
```

```python
#Total assists from each club
Assists_by_club=pd.DataFrame(epl.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
buz=sns.barplot(x='Club',y='Assists',data=Assists_by_club.sort_values(by='Assists'),palette='Set2')
buz.set_xlabel('Club',fontsize=30)
buz.set_ylabel('Assists',fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title("Plot of club vs Total assists",fontsize=20)
```

```python
# Most goal by Players top15
top_10_goals=epl[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=15,columns="Goals")
top_10_goals
```

```python
# Top 10 goals per match scorrers
top_10_gpm=epl[['Name','GoalsPerMatch','Matches','Goals']].nlargest(n=10,columns="GoalsPerMatch")
top_10_gpm
```

```python
#Top 10 players with most yellow cards
epl_yellow=epl.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Foul_Play-Yellow_card")
c=sns.barplot(x=epl_yellow['Name'],y=epl_yellow['Yellow_Cards'],label='Players',color='Yellow')
plt.ylabel('No of Yellow Cards')
c.set_xticklabels(c.get_xticklabels())
c.set_xticklabels(c.get_xticklabels(),rotation=45)
```

```python
#Top 10 players with most yellow cards
epl_red=epl.sort_values(by='Red_Cards',ascending=False)[:10]
plt.figure(figsize=(10,5))
plt.title("Harsh_Play-Red_card")
c=sns.barplot(x=epl_red['Name'],y=epl_red['Red_Cards'],label='Players',color='Red')
plt.ylabel('No of Red Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
```
